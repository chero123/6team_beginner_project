```python
from torch.utils.data import DataLoader
from dataset import PillDataset

def collate_fn(batch):
    return tuple(zip(*batch))

root = "/Users/apple/Downloads/프로젝트1/ai06-level1-project"

train_dataset = PillDataset(root=root, split="train")
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True,
                          num_workers=0, collate_fn=collate_fn)

for images, targets in train_loader:
    print("배치 이미지 개수:", len(images))
    print("첫 이미지 shape:", images[0].shape)
    print("첫 타겟 boxes:", targets[0]["boxes"].shape)
    print("첫 타겟 labels:", targets[0]["labels"])
    break
```
